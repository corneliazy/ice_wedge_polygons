```python
#extracting number of tiles in x and y direction
import glob

# loading data
# PLEASE INSERT CORRECT FILE PATH HERE
# img_list = glob.glob("C:/Users/Cornelia/Documents/Studium/EAGLE/2Semester/AdvancedProgramming/Project/sat_tiles_0_1_7cm/*.jpeg")
img_list = glob.glob("C:/Users/Cornelia/Documents/Studium/EAGLE/2Semester/AdvancedProgramming/Project/sat_tiles_0_1_29_9_png_3bands/*.png")
```

```python
# extracting numbers from filenames
# example filename:
# C:/Users/Cornelia/Documents/Studium/EAGLE/2Semester/AdvancedProgramming/Project/sat_tiles_0_1/samoylov_training_tile_0_0.jpeg
# "0" and "0" should be extracted

def extract(img):
    splitted = ((img.split("\\")[-1]).split(".")[0]).split("_")
    # y values in image tiles
    splitted_y = splitted[-1]
    # x values in image tiles
    splitted_x = splitted[-2]
    return splitted_x, splitted_y


x_y_list = [extract(i) for i in img_list]
x_y_list
```

```python
# extracting all x values
x_list = [i[0]for i in x_y_list]
# counting number of different values in list
x_list_diff = len(set(x_list))
x_list_diff
```

```python
# extracting all y values
y_list = [i[1]for i in x_y_list]
# counting number of different values in list
y_list_diff = len(set(y_list))
y_list_diff
```

```python
# https://stackoverflow.com/questions/30227466/combine-several-images-horizontally-with-python
# combine images with respect to number of tiles in x and y direction
from PIL import Image

# path to images that should be merged:
# class_img_list = glob.glob("C:/Users/Cornelia/Documents/Studium/EAGLE/2Semester/AdvancedProgramming/Project/predictions_3classes/*.jpeg")
class_img_list = glob.glob("C:/Users/Cornelia/Documents/Studium/EAGLE/2Semester/AdvancedProgramming/Project/predictions_0_1_29_9_png/*.png")
images = [Image.open(x) for x in class_img_list]
#images[1].size
#widths, heights = zip(*(i.size for i in images))
# as all images have the same size:
width, height = images[0].size

new_image = Image.new("RGB", (width*x_list_diff, height*y_list_diff))
# start values
x_offset = 0;
y_offset = 0;

# image dimensions
#x_dim = x_list_diff;
#y_dim = y_list_diff;
for im in images:
    print("xoff: ", x_offset, ", yoff: ", y_offset)
    new_image.paste(im, (x_offset, y_offset))
    
    #VALUE NEEDS TO BE ADJUSTED ACCORDING TO IMAGE SIZE!
    if(y_offset ==1344):
        y_offset = 0;
        x_offset += im.size[0];
    else:
        y_offset += im.size[1];
        #print("xoff: ", x_offset, ", yoff: ", y_offset)
        
# PLEASE CHANGE FILE NAME!
new_image.save("0_1_29_9_png.jpg")
    
```
